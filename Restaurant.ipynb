# Distribution of Ratings: Identify the overall distribution of restaurant ratings in the dataset.

```python
#visualize the distribution of rating
sns.histplot(df['Aggregate rating'], bins=30, kde=True)
plt.title("Distribution of Ratings")
plt.show()
```

As we can see that in above chart that Aggregate rating is 3.5

# Distribution of Costs: Understand the distribution of average costs for two people across all restaurants.

```python
#visualizing the distribution of costs for two people 
sns.histplot(df['Average Cost for two'].head(), bins=30, kde=True)
plt.title("Average Cost for two")
plt.show()
```

```python
city_count=df['City']
count=city_count.values
label=city_count.index()
fig=go.Figure(df[go.Pie(labels=label, values=count)])
fig.show()
```

```python
#countplot for the number of restaurant in each city.

sns.countplot(x='City', data=df, order=df['City'].value_counts().index)
plt.xticks(rotation=45, ha='right')
plt.title('Restaurant in each city')
plt.show()
```

# City-wise Cost Differences: Investigate how the average cost for two people varies across different cities.

```python
#boxplot to investigate how the average cost for two peope varies across different cities.
sns.boxplot(x='City', y='Average Cost for two', data=df.head(20))
plt.xticks(rotation=45, ha='right')
plt.title(' Average cost for two people varies across different cities')
plt.show()
```

# Relationships between Numerical Columns: Examine potential relationships between average cost, rating, and votes using the pairplot.

```python
# Pairplot for numerical columns
sns.pairplot(df[['Average Cost for two', 'Aggregate rating', 'Votes']].head(20))
plt.suptitle('Pairplot of Numerical Columns', y=1.02)
plt.show()
```
