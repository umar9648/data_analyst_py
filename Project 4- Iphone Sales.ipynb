# Iphone Sales Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the CSV data
df = pd.read_csv('apple_products.csv')

# Print summary info about the data
print(df.info())
print('\n')
print(df.describe())


```

```python
# Plot a histogram of prices
plt.hist(df['Sale Price'])
plt.title('Distribution of Prices')
plt.xlabel('Price (Rupees)')
plt.ylabel('Frequency')
plt.savefig('prices_hist.png')
plt.show()
```

```python
# Plot a bar chart of average rating by RAM
ratings_by_ram = df.groupby('Ram')['Star Rating'].mean()
ratings_by_ram.plot.bar()
plt.title('Average Rating by RAM')
plt.ylabel('Average Rating')
plt.savefig('rating_by_ram.png')  
plt.show()
```

```python
# Print analysis
print('Analysis Report')
print('---------------')
print('- Data contains {} product listings'.format(len(df)))
print('- Prices range from Rs. {} to Rs. {}'.format(df['Sale Price'].min(), df['Sale Price'].max()))  
print('- Average price is Rs. {}'.format(int(df['Sale Price'].mean())))
print('- Most common RAM size is {} GB'.format(df['Ram'].mode()[0]))
```
